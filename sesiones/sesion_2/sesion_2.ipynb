## Manejo de rutas con Pathlib

Normalmente, las rutas de archivos se escriben de manera absoluta, como cadenas:

```python
finanzas_path = r"C:\Users\micha\OneDrive\Python\taller-intro-python-datos\sesiones\sesion_2\finanzas_personales.xlsx"
```

¿Cuál es el problema con esta forma de definir rutas?
- "Solo funciona en mi máquina"
- Se debe manipular la cadena manualmente para cambiar la ruta

Pathlib es una librería, parte de la librería estándar, para manejar rutas de manera intuitiva.

Pathlib no utiliza cadenas, sino un objeto "Path" que tiene sus propios métodos.

```python
from pathlib import Path

# Siempre es bueno comenzar declarando la carpeta del script actual
CURRENT_DIR = Path.cwd()
print(CURRENT_DIR)
print(type(CURRENT_DIR))
```

Métodos del objeto "Path"

| Método | Definición |
|---|---|
| parent | Devuelve el directorio padre de la ruta. |
| name | Devuelve el nombre del archivo o carpeta final de la ruta, con extensión. |
| suffix | Devuelve la extensión del archivo, si tiene (ej: .txt). |
| mkdir() | Crea un nuevo directorio en la ruta especificada. |
| exists() | Verifica si la ruta (archivo o carpeta) existe. |
| is_file() | Retorna True si la ruta apunta a un archivo. |
| is_dir() | Retorna True si la ruta apunta a un directorio. |
| iterdir() | Itera sobre los elementos contenidos en un directorio. |
| rename() | Renombra o mueve el archivo/carpeta a una nueva ruta. |
| unlink() | Elimina el archivo al que apunta la ruta. |

```python
# Crear una nueva carpeta "data" dentro de la carpeta "sesión_2"
DATA_DIR = CURRENT_DIR / "data"

# Usamos mkdir
DATA_DIR.mkdir(exist_ok=True)
print(DATA_DIR.exists())
```

### Limpieza de DataFrames

Al hacer limpieza y manipulación, es buena práctica crear diferentes variables, como checkpoints, en lugar de reescribir la misma variable. Si la malogramos en el camino, debemos correr nuevamente la lectura de datos, y eso puede ser costoso.

Métodos de `DataFrame` para limpieza (3.2)

| Método | Definición |
| --- | --- |
| drop() | Elimina filas o columnas especificadas por etiqueta. Usa axis=0 para filas y axis=1 para columnas. |
| rename() | Cambia el nombre de columnas o índices, usando un diccionario {nombre_viejo: nombre_nuevo}. |
| replace() | Reemplaza valores específicos dentro del DataFrame por otros valores indicados. |
| fillna() | Rellena los valores nulos (NaN) con un valor especificado o un método (ej: media, valor fijo). |
| iloc() | Accede/modifica filas o columnas por posición entera (índice numérico). No acepta booleanos. |

Tareas:
- Borrar la columna "descripción"
- Renombrar "Unnamed: 4" a Amor
- Convertir la columna "Amor" a una variable dummy (0, 1)
- Corregir el error con la categoría "Transport"

```python
# drop: Queremos ahora borrar la columna "descripción"
df_limpio = df.drop(columns=["Descripción"])
df_limpio.columns
```

```python
# rename: Queremos renombrar "Unammed: 4" a "Amor"
# Esto lo hacemos con un diccionario
df_limpio = df_limpio.rename(columns={"Unnamed: 4" : "Amor"})
df_limpio.columns
```

```python
# replace: Queremos reemplazar "Amor" por una variable dummy: 
# 1 si es por amor, 0 si no lo es
print(df_limpio["Amor"].head(10))
```

```python
# Solo necesitamos reemplazar la columna "amor", así que eso será lo que reemplazamos
df_limpio["Amor"] = df_limpio["Amor"].replace({"Amor": 1})
print(df_limpio["Amor"].head(10))
```

```python
# fillna: Ahora a reemplazar NaN por 0
df_limpio["Amor"] = df_limpio["Amor"].fillna(0)
print(df_limpio.head(10))
```

```python
# iloc: Finalmente, a reemplazar "Transport" por "Transporte"
# Hay formas más eficientes de reemplazar esto, pero lo haremos con iloc para aprender a indexar
# Recuerda: (rows, columns)
print(df_limpio.iloc[3, 1])
df_limpio.iloc[3, 1] = "Transporte"
print(df_limpio.iloc[3, 1])
```

### Manipulación de DataFrames

Conjunto de técnicas para transformar, limpiar, filtrar, agrupar y reorganizar datos en estructuras tabulares con el objetivo de facilitar su análisis.

Es como editar una hoja de Excel hasta que los datos nos cuenten una historia clara.

Buscamos responder a las siguientes preguntas del negocio:

- ¿En qué categoría he gastado más dinero? (ranking)
- ¿En qué categoría he gastado más dinero del PRESENTE MES?
- ¿En qué mes he gastado más dinero?
- ¿Porcentaje de gastos por amor vs gastos normales?
- ¿Me pasé del presupuesto en alguna categoría en este mes?

Métodos de `DataFrame` para manipulación (3.3)

| Método | Definición |
| --- | --- |
| groupby() | Agrupa filas según los valores de una o más columnas, permitiendo aplicar funciones de agregación (sum, mean, count, etc.). |
| filtering (indexación booleana) | Selecciona filas que cumplen una condición específica, usando una máscara booleana (ej: df[df['col'] > 5]). |
| merge() | Combina dos DataFrames en uno solo, uniendo filas según columnas o índices comunes (similar a un JOIN de SQL). |

```python
# Vamos a crear una columna "mes" para que sea más sencillo manipular
df_limpio["Mes"] = df_limpio["Fecha"].dt.month
df_limpio.head()
```

```python
# Groupby: en qué categoría he gastado más dinero
# La primera variable es para agrupar, y la segunda es para calcular los números
monto_por_categoria = df_limpio.groupby("Categoría")["Monto"].sum().sort_values(ascending=False)
monto_por_categoria
```

```python
# groupby: - ¿En qué mes he gastado más dinero? (ranking)
# La primera variable es para agrupar, y la segunda es para calcular los números
monto_por_mes = df_limpio.groupby(df_limpio["Mes"])["Monto"].sum().sort_values(ascending=False)
monto_por_mes
```

```python
# filtrando data: En qué categoría he gastado más dinero del PRESENTE MES
# La primera variable es para agrupar, y la segunda es para calcular los números
df_limpio_jun = df_limpio[df_limpio["Mes"] == 6]
monto_por_categoria_jun = df_limpio_jun.groupby("Categoría")["Monto"].sum().sort_values(ascending=False)
monto_por_categoria_jun
```

```python
# Ahora añadimos el total
monto_por_categoria_jun["Total"] = monto_por_categoria_jun.sum()
monto_por_categoria_jun
```

```python
# groupby: - ¿Cuánto gasto por amor comparado con lo que no?
# La primera variable es para agrupar, y la segunda es para calcular los números
monto_por_amor = df_limpio.groupby("Amor")["Monto"].sum()
porcentaje_por_amor = monto_por_amor / monto_por_amor.sum() * 100
porcentaje_por_amor
```

```python
# merge: ¿Estoy sobre o bajo presupuesto?
# Leemos la tabla de presupuesto:
presupuesto = pd.read_excel(finanzas_path, sheet_name="Presupuesto")
presupuesto
```

```python
# Solo para junio
df_junio = df_limpio[df_limpio["Mes"] == 6]

gasto_categoria_mes = (
    df_junio
    .groupby(["Mes", "Categoría"])["Monto"]
    .sum()
    .reset_index()
)
gasto_categoria_mes
```

```python
# Hacemos merge a partir de categoría, esa será nuestra llave
comparacion = gasto_categoria_mes.merge(
    presupuesto,
    on="Categoría",
    how="left"
)
comparacion
```

```python
# Ahora verificamos que no nos hayamos pasado del presupuesto
comparacion_bool = comparacion["Monto"] > comparacion["Presupuesto"]
comparacion_bool
```

### Estadística descriptiva

Estadísticas descriptivas básicas

| Método | Definición |
| --- | --- |
| max() | Calcula el valor más alto de los valores numéricos de una columna o DataFrame. |
| mean() | Calcula la media (promedio) de los valores numéricos de una columna o DataFrame. |
| mode() | Calcula la moda, es decir, el valor que se repite con mayor frecuencia. |
| median() | Calcula la mediana, el valor central cuando los datos están ordenados. |
| std() | Calcula la desviación estándar, que mide la dispersión de los datos respecto a la media. |

Tareas:
- Max: ¿Cuál fue mi gasto más caro (sin contar casa)?
- Media: ¿Cuál es el gasto promedio cuando salgo por amor?
- Moda: ¿Cuál es la categoría en la que compro con mayor frecuencia?
- Mediana: ¿Cuál es la mediana de los gastos de entretenimiento?

```python
# Máximo: gasto más caro (sin contar casa)
df_limpio_sin_casa = df_limpio[df_limpio["Categoría"] != "Casa"]
max_gasto = df_limpio_sin_casa["Monto"].max()
max_gasto
```

```python
# Para ver toda la fila completa
df_limpio_sin_casa.nlargest(1, "Monto")
```

```python
# Media: gasto amor
df_limpio_solo_amor = df_limpio[df_limpio["Amor"] == 1]
df_limpio_solo_amor
```

```python
df_limpio_solo_amor["Monto"].mean()
```

```python
# ¿Cuál es la categoría en la que compro con mayor frecuencia?
moda_categoria = df_limpio["Categoría"].mode()[0]
```
