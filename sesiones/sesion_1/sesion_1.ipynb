### Variables

### String (cadena)

Un string es una secuencia de caracteres.
Uno de los tipos de datos más frecuentes.

```python
# Una cadena se guarda en una variable con comillas
mi_nombre = "Michael"
otro_nombre = "Kewin"
```

#### Métodos útiles de strings

Python incluye métodos incorporados para manipulaciones comunes con las cadenas de texto, como limpiarlas o dividirlas.

Los métodos más útiles:

| Method | Definition |
|---|---|
| upper() | Convierte todos los caracteres a mayúsculas. |
| lower() | Convierte todos los caracteres a minúsculas. |
| title() | Convierte la primera letra de cada palabra a mayúscula. |
| strip() | Elimina los espacios en blanco al inicio y al final de una cadena. |
| split() | Divide una cadena en una lista usando un separador especificado. |

```python
# Vamos a ver cómo funcionan algunos métodos

print(mi_nombre.lower())
print(mi_nombre.upper())
```

```python
nombre_con_espacios = "  Michael  "

print(f"Si no limpiamos el nombre, aparece que tuviera {len(nombre_con_espacios)} caracteres")
print(f"Si lo limpiamos con .strip(), correctamente dice que tiene {len(nombre_con_espacios.strip())} caracteres")
```

```python
mi_nombre_completo = "Michael Salvador Suárez Patilongo"
mi_nombre_lista = mi_nombre_completo.split(" ")
print(mi_nombre_lista)
```

#### F-strings

Permiten insertar variables o código dentro de strings con {}. Muy útil.

```python
# Cómo construir una función
# Con f-strings:
nombre = "Michael"
edad = 25
print(f"Hola, mi nombre es {nombre}, y tengo {edad} años.")
print(f"Mi nombre tiene {len(nombre)} caracteres.")
print(f"En el 2018 tenía {edad - (2026 - 2018)} años.")
```

## [Listas](https://www.freecodecamp.org/news/lists-in-python-comprehensive-guide/)
Son contenedores de objetos o variables de diferentes tipos.
Son ordenadas y mutables: 

- Ordenada: cada elemento tiene un índice único
- Mutable: Sus elementos son modificables.

| Método | Definition |
|---|---|
| sort() | Ordena la lista en orden ascendente. |
| append() | Agrega un solo elemento a una lista. Este elemento se ubicará **al final de la lista.** |
| extend() | Agrega múltiples elementos a una lista. |
| index() | Retorna la primera aparición del valor especificado. |

Para contar, se comienza desde el cero

```python
mi_lista = [3, 13, 2, 56, 76, 16]
mi_lista.sort()
mi_lista
# Trivia: ¿por qué esto devuelve None?: print(b.sort())
# Luego usa .reverse()
```

## 3. Control de flujo

El control de flujo es el diseño del orden en que se ejecuta el código. 

El código se escribe de arriba a abajo, pero hay muchas "palabras reservadas" para alterar el orden, repetir código, etc.

### For loop (iteraciones)

Un bucle permite "desempaquetar" un objeto iterable y repetir una sección de código por cada elemento que contiene.
Objetos iterables: cadenas y listas.

```python
# Recorremos cada elemento de la lista uno por uno
mi_lista = [1, 3, 5]
for valor in mi_lista:   # En cada vuelta, "valor" toma el siguiente elemento
    print(valor)         # Imprimimos el elemento actual
```

Así se ve internamente:

<img src="https://raw.githubusercontent.com/MichaelSuarez0/taller-intro-python-datos/main/img/for_loop.gif" width="700">

```python
# Vamos a contar cuántas "a" tiene mi nombre
for letra in mi_nombre:
    print(letra)
```

```python
# Vamos a tomar números de una lista y sumarles 1 a todos
numeros = [1, 2, 3, 4, 5]
for numero in numeros:
    print(numero + 1)
```

### If / else 

Una sentencia condicional (if / else) permite saltar secciones de código condicionalmente.

<img src="https://raw.githubusercontent.com/MichaelSuarez0/taller-intro-python-datos/main/img/if_else_block.jpg" width="600">

```python
# Ejemplo 1: if simple
edad = 20
if edad >= 18:
    print("Eres mayor de edad")
```

```python
# Ejemplo 2: if / else
numero = 7
if numero % 2 == 0:
    print("El número es par")
else:
    print("El número es impar")
```

```python
# Ejemplo 3: if / elif / else
nota = 75
if nota >= 90:
    print("Calificación: A")
elif nota >= 70:
    print("Calificación: B")
elif nota >= 50:
    print("Calificación: C")
else:
    print("Calificación: F")
```

Podemos combinar condicionales con iteraciones

```python
# Vamos a ver si mi nombre tiene alguna letra a
for letra in mi_nombre:
    if letra == "a":
        print("¡¡¡Encontré una 'a'!!!")
    else:
        print("Esta no es una 'a'")
```

### Funciones

Las funciones permiten reutilizar secciones de código para diferentes inputs (argumentos)

<img src="https://raw.githubusercontent.com/MichaelSuarez0/taller-intro-python-datos/main/img/functions.gif" width="500">

```python
def elevar_al_cuadrado(numero):
    print(numero * numero)

elevar_al_cuadrado(5)
elevar_al_cuadrado(6)
```
