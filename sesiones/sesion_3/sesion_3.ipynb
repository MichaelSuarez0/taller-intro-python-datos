## [Listas](https://www.freecodecamp.org/news/lists-in-python-comprehensive-guide/)
Es un contenedor de objetos o variables de diferentes tipos. Es ordenado y mutable. 

- Ordenado: cada elemento tiene un índice único
- Mutable: Sus elementos son modificables.

| Método 	| Definition 	|
|---	|---	|
| sort() | Ordena la lista en orden ascendente. |
| append() | Agrega un solo elemento a una lista. <br> Este elemento se ubicará al final de la lista. |
| extend() | Agrega múltiples elementos a una lista. |
| index() | Retorna la primera aparición del valor especificado. |

```python
b = [3, 13, 2, 56, 76, 16]
b.sort()
b
# Trivia: ¿por qué esto devuelve None?: print(b.sort())
```

Otros "tipos" de listas que podrían escuchar:
- **Tupla:** una lista inmutable, no se puede modificar luego de ser creada. Se crea con ().
- **Set:** una lista no ordenada y sin objetos duplicados. Se crea con {}.
